```python
from typing import TypedDict
```

```python
class PortfolioState(TypedDict):
    amount_huf: int
    total_value: int
```

```python
def calctotal_value(state: PortfolioState) -> PortfolioState:
    state['total_value'] = state['amount_huf'] * 3
    return state

def calc_something(state: PortfolioState) -> PortfolioState:
    state['total_value'] += 1
    return state
```

```python
from langgraph.graph import StateGraph, START, END

workflow = StateGraph(PortfolioState)

workflow.add_node("calctotal_node", calctotal_value)
workflow.add_node("calc_something_node", calc_something)

workflow.add_edge(START, "calctotal_node")
workflow.add_edge("calctotal_node", "calc_something_node")
workflow.add_edge("calc_something_node", END)

graph = workflow.compile()
```

```python
graph.invoke({
    "amount_huf": 1000,
    "total_value": 0
})
```
